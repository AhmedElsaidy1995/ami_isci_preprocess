```python
import pandas as pd
import codecs
import utils
import os , re
```

```python
def clean(word):
    unwantedwords = ["Um","um","Mm-hmm","_","\n"]
    for unwanted in unwantedwords:
        word = word.replace(unwanted, "")
    return re.sub('{.*?}', '', word)
```

```python
def read_dataset_files(dataset,version):
    df = pd.DataFrame(columns = ['meeting','speaker','words'])
    for root, dirs, files in os.walk(utils.get_project_dir()+"\\"+dataset):
        for file in files:
            if file.endswith(version):
                print(os.path.join(root, file))
                with open(os.path.join(root, file)) as f:
                    lines = f.readlines()
                    for line in lines:
                        rows = line.split("	", 8)
                        meeting = rows[0]
                        transcript = clean(rows[8])
                        speaker = rows[3]
                        if transcript != "" and transcript != " " and transcript != " .":
                            if df.shape[0] != 0:
                                if df['speaker'].iloc[-1] == speaker :
                                    df.iloc[-1, df.columns.get_loc('words')] += (" "+transcript)
                                else:    
                                    df = df.append({'meeting':meeting[0:7],'speaker': speaker,'words':transcript}, ignore_index=True)
                            else:
                                df = df.append({'meeting':meeting[0:7],'speaker': speaker,'words':transcript}, ignore_index=True)
    df.to_csv(dataset+'.csv')  
```

```python
read_dataset_files('ami',".2.da")
```
